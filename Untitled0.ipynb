```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/tel_data.csv')
df.head()
```

```python
models = {}

for column in ['y', 'y1', 'y2', 'y3', 'y4', 'y5']:
    model = Prophet()
    df_column = df[['ds', column]].rename(columns={column: 'y'})

    # Fit the model with your data
    model.fit(df_column)

    # Store the model in a dictionary
    models[column] = model

# Create a DataFrame with future dates for prediction (5 years into the future, annually)
future = models['y'].make_future_dataframe(periods=5, freq='AS')  # 5 periods for 5 years (AS stands for Annual Start frequency)

# Generate forecasts for each 'y' column
forecasts = {}

for column, model in models.items():
    forecast = model.predict(future)
    forecasts[column] = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]

# Display the forecasts for each 'y' column
for column, forecast in forecasts.items():
    print(f'\nForecast for {column}:\n{forecast.tail()}')
```

```python
import pandas as pd
from statsmodels.tsa.api import VAR
from statsmodels.tools.eval_measures import rmse
```

```python
df = pd.read_csv('/content/drive/MyDrive/tel_data.csv')
```

```python
columns_to_include = ['ds', 'y', 'y1', 'y2', 'y3', 'y4', 'y5']
df_multivariate = df[columns_to_include]

# Set 'ds' as the index
df_multivariate.set_index('ds', inplace=True)

# Fit a VAR model
model = VAR(df_multivariate)
model_fitted = model.fit()

# Forecast future values
n_periods = 5
forecast = model_fitted.forecast(df_multivariate.values[-model_fitted.k_ar:], steps=n_periods)

# Convert forecast to DataFrame
df_forecast = pd.DataFrame(forecast, columns=df_multivariate.columns)

# Display the forecast
plt.figure(figsize=(12, 6))

# Plot present values
plt.plot(df_multivariate.index, df_multivariate['y'], label='Present: y', marker='o')
plt.plot(df_multivariate.index, df_multivariate['y1'], label='Present: y1', marker='o')
# Repeat for other 'y' columns as needed

# Plot predicted values
plt.plot(df_forecast.index, df_forecast['y'], label='Predicted: y', linestyle='dashed', marker='o')
plt.plot(df_forecast.index, df_forecast['y1'], label='Predicted: y1', linestyle='dashed', marker='o')
# Repeat for other 'y' columns as needed

plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Present and Predicted Values')
plt.legend()
plt.show()
```

```python
import statsmodels.api as sm
data = {
    'ds': ["2015-03-31", "2016-03-31", "2017-03-31", "2018-03-31", "2019-03-31", "2020-03-31"],
    'y': [31.43, 30.93, 31.10, 32.34, 33.48, 34.28],
    'y1': [124104, 140840, 159395, 179358, 209848, 231378],
    'y2': [2333163, 2398644, 2435509, 2784477, 2905319, 2759173],
    'y3': [416332, 464542, 507946, 557410, 608401, 640968],
    'y4': [24434, 42510, 47949, 50204, 56887, 57910],
    'y5': [2650, 3948, 5154, 5213.2, 5545.2, 7303.8]
}

df = pd.DataFrame(data)

# Separate dependent and independent variables
y = df['y5']
X = df[['y']]

# Add a constant term for the intercept
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Display the regression results
print(model.summary())
```

```python
import pandas as pd
from prophet import Prophet

data = pd.read_csv('/content/drive/MyDrive/tel_2019.csv')

df = pd.DataFrame(data)

# Initialize Prophet with additional regressors
prophet = Prophet()
prophet.add_regressor('y1')
prophet.add_regressor('y2')
prophet.add_regressor('y3')
prophet.add_regressor('y4')
prophet.add_regressor('y5')

# Prepare the DataFrame
df_prophet = pd.DataFrame({
    'ds': pd.to_datetime(df['ds']),
    'y': df['y'],
    'y1': df['y1'],
    'y2': df['y2'],
    'y3': df['y3'],
    'y4': df['y4'],
    'y5': df['y5']
})

# Fit the model
prophet.fit(df_prophet)

# Create a DataFrame with future dates for prediction
future = prophet.make_future_dataframe(periods=2, freq='AS')  # Adjust the number of periods as needed

# Add the regressor values for the future dates
future['y1'] = df['y1']
future['y2'] = df['y2']
future['y3'] = df['y3']
future['y4'] = df['y4']
future['y5'] = df['y5']
future.at[5,'y1'] = 231378
future.at[5,'y2'] = 2759173
future.at[5,'y3'] = 640968
future.at[5,'y4'] = 57910
future.at[5,'y5'] = 7303.8
future.at[6,'y1'] = 231103
future.at[6,'y2'] = 2763006
future.at[6,'y3'] = 597206
future.at[6,'y4'] = 63445
future.at[6,'y5'] = 7813.5

# Make predictions
forecast = prophet.predict(future)

# Plot the forecast
fig = prophet.plot(forecast)
```
