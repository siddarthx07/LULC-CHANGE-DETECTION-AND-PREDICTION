COVARIANCE ANALYSIS

```python
import pandas as pd
import numpy as np

df = pd.read_csv("telangana 2015 - 2020 - Sheet1.csv")

row1 = 0

for i in range(1, 11):
  row2 = i
  selected_rows = df.iloc[[row1, row2], 1:]
  covariance_matrix = np.cov(selected_rows, rowvar=False)
  covariance_value = covariance_matrix[0, 1]
  df.at[row2, 'covariance'] = covariance_value
```

PRINCIPAL COMPONENT ANALYSIS

```python
c = df1.cov()
sns.heatmap(c,annot=True)

```

```python
eigenvalues, eigenvectors = np.linalg.eig(c)
print('Eigen values:\n', eigenvalues)
print('Eigen values Shape:', eigenvalues.shape)
print('Eigen Vector Shape:', eigenvectors.shape)
```

```python
idx = eigenvalues.argsort()[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:,idx]
explained_var = np.cumsum(eigenvalues) / np.sum(eigenvalues)
explained_var
```

```python
n_components = np.argmax(explained_var >= 0.50) + 1
n_components
```

```python
u = eigenvectors[1:,:n_components]
pca_component = pd.DataFrame(u,
                             index =["PUBLIC EXPENDITURE ON HEALTH (₹ Crore)s","SOCIAL SECTOR EXPENDITURE (₹ Crore)s","FISH PRODUCTION (LAKH TONNES)s","TOTAL FERTILITY RATEs","INFANT MORTALITY RATE (PER THOUSAND)s","LENGTH OF NATIONAL HIGHWAYS (KMS)s","PER CAPITA INCOME (in ₹)s","GROSS DOMESTIC PRODUCT (IN RS LAKH CRORE)s","GROSS STATE VALUE ADDED BY ECONOMIC ACTIVITY - CONSTRUCTION (₹ Lakh)s","inflation fuel and light(%)s"] ,
                             columns = ['PC1']
                            )
sns.heatmap(pca_component,annot=True)
```
